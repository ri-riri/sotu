```python
import matplotlib.pyplot as plt

# 対象データ
x = ["予期的UX", "瞬間的UX", "エピソード的UX", "累積的UX"]  # x軸の値
y1 = [20, 30, 50, 20]  # y軸の値
y2 = [50, 20, 30, 40]  # y軸の値

# figureを生成する
fig = plt.figure(figsize=(7, 4), dpi=100)

# axをfigureに設定する
ax = fig.add_subplot(1, 1, 1)
 
ax.plot(x, y1, "-", c="red", linewidth=1, marker='o')
ax.plot(x, y2, "-", c="blue", linewidth=1, marker='^')

# 汎用要素を表示
#ax.grid(True)  # grid表示ON
#ax.set_xlim(left="予期的UX", right="累積的UX")  # x範囲
ax.set_ylim(bottom=0, top=60)  # y範囲
#ax.set_xlabel('X')  # x軸ラベル
ax.set_ylabel('ネガティブ　　　　ポジティブ')  # y軸ラベル
ax.set_title('UXの期間モデルの感情曲線')  # グラフタイトル
ax.legend(['Line 1', 'Line 2'])  # 凡例を表示
xmin, xmax = "予期的UX","累積的UX"
plt.hlines([30], xmin, xmax, "gray", linestyles='dashed')     # hlines



fig.savefig('3-1_a.png')
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
```
